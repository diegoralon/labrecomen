# Primera Observación Json
Nos podemos dar cuenta que en el archivo de meta data en la columna belong_to_collection por ejemplo tenemos datos en formato Json por lo cual lo tenemos que cambiar a que sea un data frame que lo podamos leer

Para hacer el ajuste use de referencia este github: https://gist.github.com/mpHarm88/4f2fc162ea14bb3d459af0fcda38648b#file-part3-ipynb

Recordando que solo necesitamos las variables ID y Gender de la columna generes podemos sacarlas con la referencia del github pasado. Recordando que para poder usarlo en nuestra función necesitamos un valor numérico por lo cual tenemos que hacer un One hot encoding que tambien no lo proporciona el git pasado.

```python
movie_meta['genres'] = movie_meta['genres'].str.replace("\'","\"")
```

```python
movie_meta['genres'][0]
```

```python
import json
json_cols = ['genres']

def clean_json(x):
    "Create apply function for decoding JSON"
    return json.loads(x)

for x in json_cols:
    movie_meta[x] = movie_meta[x].apply(clean_json)
```

```python
def clean_json2(x):

    ls = []

    for y in range(len(x[0])):

        for k, v in x[0][y].items():
            
            ls.append(str(k)+ "_" +str(v))
   
    for z in range(len(ls)):

        if ls[z] not in movie_meta.columns:
            movie_meta[ls[z]] = 0
            movie_meta[ls[z]].iloc[x.name] = 1
        else:
            movie_meta[ls[z]].iloc[x.name] = 1
    return

for x in json_cols:
    movie_meta[[x]].apply(clean_json2, axis=1)
```
